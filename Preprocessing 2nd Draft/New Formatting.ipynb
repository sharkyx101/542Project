```python
# Convert numeric features to ints
columns_to_convert = ['winner_age', 'winner_ht', 'winner_seed', 'winner_rank', 'loser_age', 'loser_ht', 'loser_seed', 'loser_rank', 'minutes', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved', 'w_bpFaced', 'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced', 'winner_rank_points', 'loser_rank_points']

df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

df[columns_to_convert] = df[columns_to_convert].fillna(0)

df[columns_to_convert] = df[columns_to_convert].astype(int)
```

```python
# Get the data types for each feature
data_types = df.dtypes

data_types
```

```python
# Function to split score into winner score and loser score
def split_score(score):
    try:
        sets = score.split()
        winner_score = 0
        loser_score = 0
        
        for game_set in sets:
            games = game_set.split('-')

            if len(games) == 2:
                winner_score += int(games[0])
                loser_score += int(games[1])

        return winner_score, loser_score
    except (AttributeError, ValueError):
        return np.nan, np.nan  # score is not a string or contains unexpected values


df[['winner_score', 'loser_score']] = df['score'].apply(split_score).apply(pd.Series)

print(df[['score', 'winner_score', 'loser_score']].head())
```

```python
# Replace NaN values with 0 and convert to integers
df['winner_score'] = pd.to_numeric(df['winner_score'], errors='coerce').fillna(0).astype(int)
df['loser_score'] = pd.to_numeric(df['loser_score'], errors='coerce').fillna(0).astype(int)

df['winner_score'] = df['winner_score'].astype(int)
df['loser_score'] = df['loser_score'].astype(int)
```
