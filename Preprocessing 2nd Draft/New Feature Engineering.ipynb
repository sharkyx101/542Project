```python
df = pd.read_csv("InitialFeatures.csv")

df.dtypes
```

```python
# Lists to store new columns
player1_name = []
player1_seed = []
player1_height = []
player1_age = []
player1_rank = []
player1_entry = []
player1_hand = []
player1_ioc = []
player2_name = []
player2_seed = []
player2_height = []
player2_age = []
player2_rank = []
player2_entry = []
player2_hand = []
player2_ioc = []

# Iterate through rows
# Every even row has winner player 1 and loser player 2
# Every odd row has loser player 1 and winner player 2
# This technique keeps class distribution at 50/50 without altering any data
for index, row in df.iterrows():
    if index % 2 == 0: #even row
        player2_name.append(row['loser_name'])
        player2_seed.append(row['loser_seed'])
        player2_height.append(row['loser_ht'])
        player2_age.append(row['loser_age'])
        player2_rank.append(row['loser_rank'])
        player2_entry.append(row['loser_entry'])
        player2_hand.append(row['loser_hand'])
        player2_ioc.append(row['loser_ioc'])
        
        player1_name.append(row['winner_name'])
        player1_seed.append(row['winner_seed'])
        player1_height.append(row['winner_ht'])
        player1_age.append(row['winner_age'])
        player1_rank.append(row['winner_rank'])
        player1_entry.append(row['winner_entry'])
        player1_hand.append(row['winner_hand'])
        player1_ioc.append(row['winner_ioc'])
        
    else: #odd row
        player1_name.append(row['loser_name'])
        player1_seed.append(row['loser_seed'])
        player1_height.append(row['loser_ht'])
        player1_age.append(row['loser_age'])
        player1_rank.append(row['loser_rank'])
        player1_entry.append(row['loser_entry'])
        player1_hand.append(row['loser_hand'])
        player1_ioc.append(row['loser_ioc'])
        
        player2_name.append(row['winner_name'])
        player2_seed.append(row['winner_seed'])
        player2_height.append(row['winner_ht'])
        player2_age.append(row['winner_age'])
        player2_rank.append(row['winner_rank'])
        player2_entry.append(row['winner_entry'])
        player2_hand.append(row['winner_hand'])
        player2_ioc.append(row['winner_ioc'])

# Create new columns
df['player1'] = player1_name
df['player1_age'] = player1_age
df['player1_height'] = player1_height
df['player1_hand'] = player1_hand
df['player1_ioc'] = player1_ioc
df['player1_rank'] = player1_rank
df['player1_seed'] = player1_seed
df['player1_entry'] = player1_entry

df['player2'] = player2_name
df['player2_age'] = player2_age
df['player2_height'] = player2_height
df['player2_hand'] = player2_hand
df['player2_ioc'] = player2_ioc
df['player2_rank'] = player2_rank
df['player2_seed'] = player2_seed
df['player2_entry'] = player2_entry

# Create a winner column that indicates who won 
#1 for player1
#2 for player2
df['winner'] = df.index % 2 + 1

# Drop unnecessary columns
df.drop(['winner_name', 'loser_name', 'winner_seed', 'winner_ht', 'winner_age', 'winner_rank', 'winner_entry', 'winner_hand', 'winner_ioc', 'loser_seed', 'loser_ht', 'loser_age', 'loser_rank', 'loser_entry', 'loser_hand', 'loser_ioc'], axis=1, inplace=True)

# Display the updated DataFrame
print(df.head())
```
