```python
#Michael Sanati
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("InitialFormatting.csv")
```

```python
sns.scatterplot(x="winner_ht", y="loser_ht", data=df)
plt.xlabel("Winner's Height")
plt.ylabel("Loser's Height")
plt.title("Scatter Plot of Winner's Height vs Loser's Height")
plt.show()
```
